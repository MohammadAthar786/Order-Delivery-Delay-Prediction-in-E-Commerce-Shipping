# 🧠 Python venv Cheatsheet

---

# 📦 1. Create Environment

```bash
python -m venv myenv
```

👉 Folder banega:

```text
myenv/
```

---

# ▶️ 2. Activate Environment

### 🪟 Windows:

```bash
myenv\Scripts\activate
```

### 🍎 Mac/Linux:

```bash
source myenv/bin/activate
```

👉 Activate hone par:

```bash
(myenv) C:\...
```

---

# ❌ 3. Deactivate Environment

```bash
deactivate
```

---

# 📍 4. Check Current Python

```bash
where python   # Windows
```

👉 Ya notebook me:

```python
import sys
print(sys.executable)
```

---

# 📥 5. Install Packages

```bash
pip install pandas numpy
```

---

# 📋 6. List Installed Packages

```bash
pip list
```

---

# 📄 7. Save Requirements

```bash
pip freeze > requirements.txt
```

---

# 📦 8. Install from Requirements

```bash
pip install -r requirements.txt
```

---

# 🧹 9. Delete Environment

👉 Simply folder delete karo:

```bash
rm -rf myenv   # Mac/Linux
```

```bash
rmdir /s myenv   # Windows
```

---

# 💥 Golden Rule

👉 **"Activate env → then install → then run"**

---
📁 1. Ideal Project Structure
```
my_project/
│
├── myenv/                # virtual environment (ignore in git)
├── src/                  # main code
│   ├── __init__.py
│   ├── main.py
│   └── utils.py
│
├── notebooks/            # Jupyter notebooks
│   └── analysis.ipynb
│
├── requirements.txt      # dependencies
├── .gitignore
├── README.md
└── config.py (optional)
```
🧠 Why this structure?\
src/ → clean production code\
notebooks/ → experiments alag\
myenv/ → isolate dependencies\
requirements.txt → reproducibility

.gitignore (VERY IMPORTANT)
```
myenv/
__pycache__/
*.pyc
.ipynb_checkpoints/
```

👉 2 types ke kernel usage:

Permanent kernel\
ipykernel install se banta hai\
kernelspec list me dikhta hai\
Temporary kernel (VS Code)\
auto create hota hai\
list me nahi dikhega



## 🔹 2. `where python` vs `where pip`

### ✅ `where python`

* System mein **all python paths** dikhata hai
* Top = default python

if environment is activated then top path is the python path that environnment is using
---

### ✅ `where pip`

* **pip kis python se linked hai** dikhata hai

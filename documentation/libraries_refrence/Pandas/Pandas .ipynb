```python
🔹 Pandas Series (Single Column Data)
Series ek single column hoti hai jisme index automatically set hota hai.

✅ Example:
```

```python
import pandas as pd


data = [10, 20, 30, 40]
series = pd.Series(data)
print(series)

```

```python
#👉 Yahan 0, 1, 2, 3 index values hain. Hum chahein toh custom index bhi de sakte hain:
series = pd.Series(data, index=['a', 'b', 'c', 'd'])
print(series)

```

```python
# 🔹 Pandas DataFrame (Table Format Data)
# DataFrame ek table hoti hai jo multiple columns aur rows ko store karti hai.


data = {
    'Name': ['Ali', 'Sara', 'Ahmed'],
    'Age': [25, 23, 30],
    'City': ['Delhi', 'Mumbai', 'Chennai']
}

df = pd.DataFrame(data)
print(df)
# 👉 Yahan humne ek dictionary se DataFrame banaya hai.
```

```python

# Data Science me aksar hum CSV (Excel jaisa format) files use karte hain.

# CSV (Comma-Separated Values) ek simple text file format hai jo structured data ko store karne ke liye use hota hai. Is file format mein data rows aur columns ki form mein hota hai, jismein har value comma (,) se separate hoti hai.

# CSV File ki Key Features:
# Simple Structure: Har row ek naya record hota hai, aur har column ek attribute ya field ko represent karta hai.
# Comma-Separated Values: Columns ko alag karne ke liye commas (,) ka istemal hota hai.
# Plain Text Format: Ye ek normal text file hoti hai jise kisi bhi text editor (Notepad, VS Code) ya spreadsheet software (MS Excel, Google Sheets) mein khola ja sakta hai.
# Lightweight & Fast Processing: JSON ya XML ki comparison mein CSV files lightweight hoti hain aur easily process ho sakti hain.
# Platform-Independent: Is format ko har programming language (Python, C++, Java) easily handle kar sakti hai.
# CSV File Example:
# Agar ek CSV file students.csv hai, to usmein data kuch is tarah se hoga:
# ID,Name,Age,Marks
# 1,Ali,22,85
# 2,Ahmed,21,90
# 3,Aisha,23,88
# Yahan:
# ID ek column hai jo student ka unique number batata hai.
# Name student ka naam hai.
# Age student ki umar hai.
# Marks uske obtained marks hain.
# CSV File ka Use:
# Data Storage: Tabular data store karne ke liye.
# Data Exchange: Different applications ke beech data share karne ke liye.
# Data Processing: Python ya Excel ke through analysis aur visualization ke liye.
```

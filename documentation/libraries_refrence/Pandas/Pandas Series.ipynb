# Important `Series` Methods in Pandas

## 1. Basic Information and Statistics
| Method | Description |
|--------|-------------|
| `s.head(n)` | Returns the first `n` elements of the Series. |
| `s.tail(n)` | Returns the last `n` elements of the Series. |
| `s.info()` | Displays basic information about the Series. |
| `s.describe()` | Provides summary statistics (count, mean, std, min, max, etc.). |
| `s.value_counts()` | Returns unique values and their frequencies. |
| `s.count()` | Counts non-null values. |
| `s.mean(), s.median(), s.mode()` | Computes the mean, median, and mode of the Series. |
| `s.min(), s.max(), s.std()` | Returns the min, max, and standard deviation. |

---


```python

import pandas as pd

s = pd.Series([10, 20, 10, 30, 40, 50, 10])

print(s.head(3))  # First 3 elements
print(s.describe())  # Summary statistics
print(s.value_counts())  # Unique values and their counts
```

## 2. Indexing & Selection

| Method | Description |
|--------|-------------|
| `s.loc[label]` | Access elements using labels. |
| `s.iloc[index]` | Access elements using index position. |
| `s.at[label]` | Fast access for a single label. |
| `s.iat[index]` | Fast access for a single index. |


```python
s = pd.Series([100, 200, 300], index=['a', 'b', 'c'])

print(s.loc['b'])  # Access element by label
print(s.iloc[1])   # Access element by index
```


## 3. Data Transformation
| Method | Description |
|--------|-------------|
| `s.apply(func)` | Applies a function element-wise. |
| `s.map(func)` | Maps values using a function or dictionary. |
| `s.astype(dtype)` | Converts the Series to a specific data type. |
| `s.replace(old, new)` | Replaces values in the Series. |
| `s.clip(lower, upper)` | Limits values to a specified range. |
| `s.round(n)` | Rounds values to `n` decimal places. |

---


```python
s = pd.Series([1.234, 5.678, 9.012])

print(s.round(1))  # Round to 1 decimal place
print(s.map(lambda x: x * 2))  # Multiply each value by 2
```


## 4. Handling Missing Values
| Method | Description |
|--------|-------------|
| `s.isna()` | Returns a Boolean Series indicating missing values. |
| `s.notna()` | Returns a Boolean Series indicating non-missing values. |
| `s.fillna(value)` | Replaces NaN values with a specified value. |
| `s.dropna()` | Removes missing values. |

---


```python
s = pd.Series([1, 2, None, 4, None, 6])

print(s.fillna(0))  # Replace NaN with 0
print(s.dropna())   # Remove NaN values
```

# Ranking Concept
In Pandas, rank refers to the relative position of each value in a Series when sorted in ascending or descending order.

For example, in a given list of numbers, the smallest number gets rank 1, the second smallest gets rank 2, and so on. If there are duplicate values, the ranking method determines how their ranks are assigned.

# Example of Ranking Concept
[10, 20, 15, 10, 30]

The rank of each element in ascending order (smallest to largest) would be: 

10 → 1.5  
10 → 1.5  
15 → 3  
20 → 4  
30 → 5

Here, 10 appears twice, so it gets an average rank of (1+2)/2 = 1.5.\
Why Use Ranking?\
Ranking is useful in:

Data analysis (e.g., ranking students by scores)\
Sorting data without modifying original order\
Handling ties systematically with different ranking methods

```python
s = pd.Series([50, 10, 30, 40])

print(s.sort_values())  # Sort by values
print(s.rank())  # Compute rank
```

# Rolling Mean with a Window of Size N
The rolling mean (also called moving average) is a technique in time series analysis where we calculate the average of the last
N values in a sequence. It helps smooth out fluctuations in data, making trends easier to see.

Concept:

A rolling mean with a window of size N means that for each point in the dataset, we take the previous N values (including the current one) and compute their average.


# Example: Rolling Mean with Window Size = 3
Step 1: Given Data
Let’s say we have the following Pandas Series representing daily stock prices: 

 Day:   1   2   3   4   5   6
Price: 10  20  30  40  50  60
 
Step 2: Compute Rolling Mean (Window = 3) 

For a window size of 3, we compute the mean for every three consecutive values:
First 3 values: (10 + 20 + 30) / 3 = 20.0  
Next 3 values:  (20 + 30 + 40) / 3 = 30.0  
Next 3 values:  (30 + 40 + 50) / 3 = 40.0  
Next 3 values:  (40 + 50 + 60) / 3 = 50.0  

Since we need at least 3 values to compute the rolling mean, the first two values will be NaN (not available).




```python
# Step 3: Implementation in Pandas
import pandas as pd

# Create a Pandas Series
data = pd.Series([10, 20, 30, 40, 50, 60])

# Compute rolling mean with window size 3
rolling_mean = data.rolling(window=3).mean()

print(rolling_mean)
```

First two values are NaN because there aren’t enough values to form a window of size 3.
The rolling mean starts at index 2 when we have enough values to compute it.

Why Use Rolling Mean?

Smooths fluctuations in noisy data.\
Identifies trends in stock prices, weather patterns, etc.\
Removes short-term variations to focus on long-term trends.

# Real World Example of Rolling mean and How it is used 

Example: Rolling Mean on Stock Prices\
Step 1: Create a DataFrame\
We'll simulate daily closing prices of a stock and calculate the 7-day rolling mean.


```python
import pandas as pd
import numpy as np

# Creating a DataFrame with stock prices (simulated)
data = {
    "Date": pd.date_range(start="2024-01-01", periods=10, freq="D"),
    "Stock_Price": [100, 102, 105, 103, 110, 108, 107, 112, 115, 118]
}

# pd.date_range() generates a sequence of dates over a specified range.
"""
pd.date_range(start, end, periods, freq)

start="YYYY-MM-DD" → Defines the starting date of the range.
end="YYYY-MM-DD" → (Optional) Defines the ending date.
periods=N → Specifies how many dates to generate.
freq="D" → Defines the frequency of dates (daily in this case).

pd.date_range(start="2024-01-01", periods=10, freq="D")

Generated Date Range Output:

DatetimeIndex(['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04',
               '2024-01-05', '2024-01-06', '2024-01-07', '2024-01-08',
               '2024-01-09', '2024-01-10'],
              dtype='datetime64[ns]', freq='D')
Other Examples
Example 1: Weekly Frequency (freq="W")
pd.date_range(start="2024-01-01", periods=5, freq="W")

output:

DatetimeIndex(['2024-01-07', '2024-01-14', '2024-01-21', '2024-01-28', '2024-02-04'],
              dtype='datetime64[ns]', freq='W-SUN')
* Generates 5 weekly dates, starting from January 7, 2024 (default: Sundays).

Example 2: Hourly Frequency (freq="H")
pd.date_range(start="2024-01-01", periods=5, freq="H")

DatetimeIndex(['2024-01-01 00:00:00', '2024-01-01 01:00:00', 
               '2024-01-01 02:00:00', '2024-01-01 03:00:00', 
               '2024-01-01 04:00:00'],
              dtype='datetime64[ns]', freq='H')
Example 3: Monthly Frequency (freq="M")
pd.date_range(start="2024-01-01", periods=5, freq="M")

output:
DatetimeIndex(['2024-01-31', '2024-02-29', '2024-03-31', '2024-04-30', '2024-05-31'],
              dtype='datetime64[ns]', freq='M')

*Returns the last day of each month.

Use Cases
1 Creating time-based datasets for stock prices, temperature logs, etc.
2 Simulating future dates for predictions or analysis.
3 Generating time intervals for scheduling tasks.

"""

df = pd.DataFrame(data)

# Compute 7-day rolling mean
df["7-day Rolling Mean"] = df["Stock_Price"].rolling(window=7).mean()

print(df)
```

# Explanation
The 7-day rolling mean is the average of the past 7 days (including the current day).\
The first 6 values are NaN because we need at least 7 values to compute the mean.\
Once we reach day 7, we start getting values.

Why Is This Useful?


*Stock Market Analysis: Helps identify trends by reducing daily fluctuations.\
*Weather Forecasting: Used to smooth temperature data.\
*Financial Metrics: Commonly used for Moving Averages (e.g., 50-day, 200-day MA).

````python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head()
````

# ✅ Step 4: Bivariate & Relationship Plots
## 🎯 Goal:
### Explore relation between two numeric features.

#### 📊 Plots:
| Plot            | Use                               |
| --------------- | --------------------------------- |
| `scatterplot()` | 2D relation                       |
| `lineplot()`    | Trend lines                       |
| `regplot()`     | Scatter + regression              |
| `lmplot()`      | Multiple linear plots by category |
| `relplot()`     | Faceted version of scatter/line   |


# Scatter Plot
➤ Used to visualize the relationship between two numerical variables, optionally grouped by categories (hue) or other dimensions like size and style.
```
sns.scatterplot(x=None, y=None, data=None, hue=None, size=None, style=None, ...)
```
main Parameters

| Parameter | Description                                                    |
| --------- | -------------------------------------------------------------- |
| `x` / `y` | Numerical variables for horizontal/vertical axes               |
| `data`    | DataFrame containing the variables                             |
| `hue`     | Categorical variable for color grouping                        |
| `size`    | Numerical or categorical variable for point size               |
| `style`   | Categorical variable for marker style (e.g., circle, triangle) |
| `palette` | Color palette                                                  |
| `sizes`   | Range of sizes for `size` variable                             |
| `markers` | Manual marker styles                                           |
| `alpha`   | Transparency (0 = invisible, 1 = solid)                        |
| `legend`  | Show/hide legend (`"full"` or `False`)                         |


````python
# Basics Example

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bill vs Tip")
plt.show()
````

````python
# CUSTOMIZATION
#🔸 Add Hue (color by category)
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
````

````python
# 🔸 Add Size Variation
sns.scatterplot(x="total_bill", y="tip", size="size", data=tips)
````

````python
# 🔸 Add Style Variation (marker shape)
sns.scatterplot(x="total_bill", y="tip", style="sex", data=tips)
````

````python
# 🔸 Combine Hue + Style + Size
sns.scatterplot(x="total_bill", y="tip", hue="day", style="sex", size="size", data=tips)
````

````python
# 🔸 Add Transparency
sns.scatterplot(x="total_bill", y="tip", data=tips, alpha=0.6)
````

````python
# 🔍 6. Bonus: Add Trend Line
# If you want to show the regression line, use sns.lmplot() or sns.regplot() instead:
sns.regplot(x="total_bill", y="tip", data=tips)
````

# Line Plot
➤ Used to visualize the trend/relationship between two variables, typically when the x-axis is time-based or ordered.
```
sns.lineplot(x=None, y=None, data=None, hue=None, style=None, markers=False, dashes=True, ...)
```
| Parameter   | Description                                                  |
| ----------- | ------------------------------------------------------------ |
| `x`, `y`    | Variables to plot (usually numerical or ordered)             |
| `data`      | DataFrame                                                    |
| `hue`       | Categorical variable for line color                          |
| `style`     | Variable for line style (dashed, dotted, etc.)               |
| `size`      | Line thickness variation                                     |
| `palette`   | Custom color palette                                         |
| `markers`   | Add markers at data points (`True`, `False`, or dict)        |
| `dashes`    | Use dashed lines for categories (`True`, `False`, or dict)   |
| `err_style` | Style of error bars: `'band'`, `'bars'`, or `None`           |
| `ci`        | Confidence interval (default: 95%). Set `ci=None` to disable |

## Interpretation Tips
| Feature        | Meaning                                   |
| -------------- | ----------------------------------------- |
| Slope ↑        | Increasing trend                          |
| Slope ↓        | Decreasing trend                          |
| Flat Line      | No change over x-axis                     |
| Multiple lines | Compare different groups over same x-axis |



````python
# ✅  Basic Example



flights = sns.load_dataset("flights")
print(flights.head())

sns.lineplot(x="year", y="passengers", data=flights)
plt.title("Total Passengers Per Year")
plt.ylabel("Number of Passengers")
plt.show()
````

````python
sns.lineplot(x="size", y="tip", data=tips)
plt.title("Size vs Tip")
````

````python
# 🎨 4. Customizations
# 🔸 Multiple Lines with Hue
sns.lineplot(x="year", y="passengers", hue="month", data=flights)
plt.title("Monthly Passenger Trends Across Years")
plt.ylabel("Number of Passengers")
plt.show()
# Each line shows how a month behaves across years (e.g., July consistently peaks).
````

````python
# 🔸 Add Style (dashed vs solid)
sns.lineplot(x="year", y="passengers", hue="month", style="month", data=flights)
````

````python
# 🔸 Add Markers
sns.lineplot( x="year", y="passengers", hue="month", style="month", markers=True, data=flights)
````

````python
# 🔸 Disable Confidence Interval (CI)

sns.lineplot( x="year", y="passengers", hue="month", style="month",ci=None,data=flights)
````

````python
# 🔸 Change Color Palette
sns.lineplot(x="year", y="passengers", data=flights, palette="Set2")
````

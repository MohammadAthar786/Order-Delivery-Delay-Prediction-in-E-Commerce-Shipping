# Histogram 

✅ Purpose
Histogram is used to visualize the distribution of a single numerical variable by placing values into bins and counting how many fall into each bin.

It’s the go-to plot for:

Detecting skewness, outliers, and spread

Understanding data distribution before modeling

```
sns.histplot(data=None, x=None, y=None, hue=None, bins='auto', kde=False, stat='count', multiple='layer', element='bars', fill=True)
```

| Argument   | Description                                                                            |
| ---------- | -------------------------------------------------------------------------------------- |
| `data`     | DataFrame source                                                                       |
| `x`, `y`   | Variable to plot (only 1 required, usually `x`)                                        |
| `hue`      | Separate color groups (categorical feature)                                            |
| `bins`     | Number of bins or bin edges (e.g., 10 or list of edges)                                |
| `kde`      | `True` overlays smooth kernel density estimate                                         |
| `stat`     | `"count"` (default), `"frequency"`, `"probability"`, `"density"`, or `"percent"`       |
| `element`  | `"bars"` (default), `"step"`, or `"poly"` style                                        |
| `multiple` | How multiple hue levels are shown: `"layer"` (default), `"stack"`, `"dodge"`, `"fill"` |
| `fill`     | Fill the bars (`True`) or just outline (`False`)                                       |


````python
# 📘 Examples
# 🔹 1. Basic Histogram
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")

sns.histplot(data=tips, x="total_bill")
````

````python
# 🔹 2. With KDE Line
sns.histplot(data=tips, x="total_bill", kde=True)

````

````python
# 🔹 3. Custom Number of Bins
sns.histplot(data=tips, x="total_bill", bins=20)
````

````python
# 🔹 4. Histogram with Categorical Hue
sns.histplot(data=tips, x="total_bill", hue="sex", multiple="stack")
````

````python
# 🔹 5. Horizontal Histogram
sns.histplot(data=tips, y="total_bill", hue="sex")
````

````python
# 🔹 6. Histogram with Density Instead of Count
sns.histplot(data=tips, x="total_bill", stat="density", kde=True)
````

## Interpretation Guide
| Feature                   | Meaning                                      |
| ------------------------- | -------------------------------------------- |
| Bell curve shape          | Normally distributed                         |
| Long tail (right or left) | Skewed data                                  |
| Tall bar(s) at one end    | Many small or large values (outliers)        |
| Gaps or multiple peaks    | Possible clusters or multimodal distribution |

### When to use Histogram
| Use Case                                   | Plot                 |
| ------------------------------------------ | -------------------- |
| Distribution of a single numeric feature   | Histogram            |
| Comparing distribution across a category   | Histogram with `hue` |
| Checking for skewness                      | Histogram + KDE      |
| Detecting modality (1 peak, 2 peaks, etc.) | Histogram            |


## KDE Plot
A KDE plot is a smoothed version of a histogram used to estimate the probability density function (PDF) of a continuous variable.
It helps visualize where values are concentrated in the distribution.
```
✅ Syntax: sns.kdeplot()
sns.kdeplot(
    data=None,
    x=None,
    y=None,
    fill=False,
    bw_adjust=1,
    hue=None,
    multiple='layer',
    common_norm=True,
    palette=None,
    alpha=...,
    linewidth=1.5
)
```
| Parameter     | Purpose                                                         |
| ------------- | --------------------------------------------------------------- |
| `data`        | Dataset                                                         |
| `x`, `y`      | Column names (x for univariate, x & y for bivariate)            |
| `fill=True`   | Fills area under the curve                                      |
| `bw_adjust`   | Bandwidth control (smoothing) — higher = smoother               |
| `hue`         | Group KDEs by category (e.g. gender)                            |
| `multiple`    | How KDEs are layered: `'stack'`, `'fill'`, `'layer'`, `'dodge'` |
| `common_norm` | Normalize all curves together or separately                     |
| `cut`         | How far to extend curve beyond data                             |


````python
# 📌 Basic Example:
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

tips = load_dataset("tips")

sns.kdeplot(data=tips, x="total_bill")
plt.title("KDE of Total Bill")
plt.show()
````

````python
# 📊 Example with Fill and Hue:
sns.kdeplot(data=tips, x="total_bill", fill=True, hue="sex", alpha=0.4, linewidth=2)
plt.title("KDE of Total Bill by Gender")
````

````python
# 🧠 Bandwidth Adjustment (bw_adjust)
# Less smooth (smaller bandwidth)
sns.kdeplot(data=tips, x="total_bill", bw_adjust=0.3)

# Very smooth (larger bandwidth)
sns.kdeplot(data=tips, x="total_bill", bw_adjust=2)
````

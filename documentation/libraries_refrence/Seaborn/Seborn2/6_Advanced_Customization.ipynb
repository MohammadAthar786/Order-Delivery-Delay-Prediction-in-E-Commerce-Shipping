# ✅  Advanced Customization & Themes

🎯 Goal:\
Make your plots publication-quality.

🧰 Tools:\
sns.set_style("whitegrid")

palette="coolwarm" or custom palettes

context='talk' for larger fonts

hue, size, style for aesthetics

Axis labels, legends, titles

```python
# ✅ 1. sns.set_theme() – Set Overall Style

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
tips = sns.load_dataset("tips")

# Set a clean whitegrid theme and context for better visibility
sns.set_theme(style="whitegrid", context="talk")
# style= options: "whitegrid" (default), "darkgrid", "white", "dark", "ticks"
# context= options: "notebook", "talk", "paper", "poster"
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day", fontsize=16)
plt.show()
```

```python
# ✅ 2. Control Figure Size, Labels, Ticks
plt.figure(figsize=(10, 5))
sns.barplot(x="day", y="total_bill", data=tips)

plt.title("Average Total Bill per Day", fontsize=18, color="purple")
plt.xlabel("Day", fontsize=14)
plt.ylabel("Total Bill", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
# ✅ 3. Color Palettes
# 📌 Example 1: Built-in Palettes
sns.set_palette("pastel")
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Distribution of Total Bill by Day")
plt.show()

#📌 Example 2: Custom Palette
my_colors = ["#FF6F61", "#6B5B95", "#88B04B", "#F7CAC9"]
sns.set_palette(my_colors)

sns.countplot(x="day", data=tips)
plt.title("Count of Records by Day")
plt.show()
```

```python
# ✅ 4. Annotations & Lines
avg = tips["total_bill"].mean()

sns.histplot(tips["total_bill"], kde=True)
plt.axvline(avg, color="red", linestyle="--", label=f"Average: {avg:.2f}")
# we can draw also Horizantal line using axhline
plt.legend()
plt.title("Total Bill Distribution with Mean Line")
plt.show()
```

```python
# ✅ 5. Subplots (Multiple Charts Side by Side)
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(x="sex", y="total_bill", data=tips, ax=axs[0])
axs[0].set_title("Boxplot by Sex")

sns.violinplot(x="sex", y="total_bill", data=tips, ax=axs[1])
axs[1].set_title("Violinplot by Sex")

plt.suptitle("Comparison of Box and Violin Plot", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# ✅ 6. Save Plot to File
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bill vs Tip")
plt.savefig("scatter_plot.jpg", dpi=300, bbox_inches="tight")
# ✅ Always use dpi=300 for publication or PDF reports.
```

```python
# ✅ 7. Fonts & Styles
plt.rcParams["font.family"] = "Comic Sans MS"  # or 'monospace', 'Comic Sans MS'

sns.lineplot(x="size", y="tip", data=tips)
plt.title("Lineplot Example with Custom Font")
plt.show()
```

## ✅ Summary Table

| What You Want    | How to Do It                                     |
| ---------------- | ------------------------------------------------ |
| Theme            | `sns.set_theme(style, context)`                  |
| Font size & type | `plt.rcParams['font.family']`, `fontsize=`       |
| Custom colors    | `sns.set_palette()`                              |
| Rotate labels    | `plt.xticks(rotation=45)`                        |
| Draw lines       | `plt.axvline()` or `plt.axhline()`               |
| Add annotation   | `plt.text(x, y, s='Note')`                       |
| Save plot        | `plt.savefig("plot.png", dpi=300)`               |
| Create subplots  | `fig, axs = plt.subplots()` and pass `ax=axs[i]` |

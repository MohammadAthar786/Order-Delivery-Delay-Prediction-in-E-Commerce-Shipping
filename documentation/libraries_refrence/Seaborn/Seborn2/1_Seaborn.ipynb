```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Load sample datasets for practice:
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
print("tips dataset coumns: ",tips.columns)
print("iris dataset columns :",iris.columns)
num_feature_tip_dataset=tips.select_dtypes(include=['float','int'])
cat_features_tips_dataset=tips.select_dtypes(include=['object','category'])
num_feature_iris_dataset = iris.select_dtypes(include=["float", "int"])
cat_features_iris_dataset = iris.select_dtypes(include=["object", "category"])

print("numerical features in tips dataset ",num_feature_tip_dataset.columns)
print("categorical features are in tips dataset : ",cat_features_tips_dataset.columns)

print("numerical features in iris dataset ", num_feature_iris_dataset.columns)
print("categorical features are in iris dataset : ", cat_features_iris_dataset.columns)
```

```python
# separating dataset into target and independent features for tips dataset 
target = "tip"
print("target feature in tips dataset : ", target)
x = tips.drop(columns='tip')
y = tips['tip']
print("independent features in tips dataset: ", list(x.columns))
print("dependent features in tips dataset: ",y.name)
print(x.shape)
print(y.shape)
```

```python
# separating dataset into target and independent features for iris  dataset
target = 'species'
print("target feature in iris dataset : ", target)
x = iris.drop(columns="species")
y = iris[target]
print("independent features in iris  dataset: ", list(x.columns))
print("dependent features in iris dataset: ", y.name)
print(x.shape)
print(y.shape)
```

```python
# Core Plotting Functions
# Distribution Plots
# Histogram:
sns.histplot(tips)
```

```python
# KDE Plot:
sns.kdeplot(iris)
```

```python
# Box Plot:
## in x we pass categorical features and in y we pass numerical feature to show distribution
sns.boxplot(x='sex', y='total_bill',data=tips)
```

```python
sns.boxplot(x="smoker", y="size", data=tips)
```

```python
# Violin Plot
sns.violinplot(x='species', y='petal_length', data=iris)
```

```python
# Categorical Plots
## Bar Plot:
sns.barplot(x='species', y='petal_width', data=iris)
```

```python
sns.countplot(x='sex',data=tips)
```

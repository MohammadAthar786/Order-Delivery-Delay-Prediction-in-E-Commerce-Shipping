````python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head()
````

# ✅ Step 3: Categorical Plots (Univariate + Category)

## Visualize how one categorical and one numerical feature relate.
### 📊 Plots:
| Plot           | Description                  |
| -------------- | ---------------------------- |
| `barplot()`    | Show mean (with CI)          |
| `countplot()`  | Count of categories          |
| `boxplot()`    | Show distribution & outliers |
| `violinplot()` | KDE + Boxplot                |
| `stripplot()`  | Scatter of individual points |

# Bar Plot
✅ Purpose:
Show mean (or estimator) of numeric variable across categories.
```
🔧 Syntax: sns.barplot(x=None, y=None, data=None, estimator=np.mean, ci=95,.......)
```
| Parameter   | Description                                                         |
| ----------- | ------------------------------------------------------------------- |
| `x`         | Categorical variable (grouping axis)                                |
| `y`         | Numerical variable (height of bars)                                 |
| `data`      | DataFrame containing the variables                                  |
| `estimator` | Function used to aggregate data (default: `np.mean`)                |
| `ci`        | Size of confidence interval (default: 95). Use `ci=None` to remove. |
| `palette`   | Color palette (e.g., `"Blues"`, `["red", "blue"]`, or `hue`-based)  |
| `hue`       | Additional categorical variable for subgroups (side-by-side bars)   |
| `order`     | Custom order of categories                                          |
| `capsize`   | Adds caps to error bars                                             |
| `errcolor`  | Color of the error bars                                             |
| `linewidth` | Width of the bar borders                                            |

                        

🧠 Interpretation:

Height of bar = average value.(if used default np.mean)

Error bars = confidence interval.

````python
# default use case
sns.barplot(x="sex", y="tip", data=tips)
````

````python
# CUSTOMIZATION
# Change Aggregation with Different Function
import numpy as np

# Show the sum of total_bill by day
sns.barplot(x="day", y="total_bill", data=tips, estimator=np.sum)
````

````python
# changing orientation
sns.barplot(x="total_bill", y="day", data=tips, orient="h")
````

````python
# Adding hue -categorized distribution
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
````

# Box plot
#### Used to show distribution, central tendency, and spread of a numerical variable, often grouped by categories.
```
sns.boxplot(x=None, y=None, data=None, hue=None, ...)
```
| Parameter   | Description                                    |
| ----------- | ---------------------------------------------- |
| `x`         | Categorical variable (used for grouping)       |
| `y`         | Numerical variable (used for plotting boxplot) |
| `data`      | DataFrame containing the variables             |
| `hue`       | Subgrouping within each category               |
| `order`     | Custom order of categories                     |
| `palette`   | Color scheme                                   |
| `orient`    | Orientation: `'v'` or `'h'`                    |
| `width`     | Width of box (default is 0.8)                  |
| `fliersize` | Size of outlier points                         |
| `linewidth` | Width of box and whisker lines                 |
| `notch`     | If True, draw notched box                      |

### Each box represents:

Median (middle line in box)

Q1–Q3 (Interquartile Range, IQR) — the box itself

Whiskers — extend to 1.5 × IQR

Outliers — dots outside whiskers

### 🔍 4. Interpretation
Taller boxes → more spread in data.

Outliers → dots beyond whiskers.

Position of the median line inside the box shows skewness:

Near top → left-skewed

Near bottom → right-skewed



````python
# Basic Example
sns.boxplot(x="day", y="total_bill", data=tips)

plt.title("Total Bill Distribution by Day")
plt.show()
````

````python
# 🎨 CUSTOMIZATION
##🔸 Add hue for subgroups (e.g., gender within day)
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
````

````python
# 🔸 Horizontal Boxplot
sns.boxplot(x="total_bill", y="day", data=tips)
````

````python
# 🔸 Remove outliers
sns.boxplot(x="day", y="total_bill", data=tips, showfliers=False)
````

````python
# 🔸 Notched Box (shows confidence interval around median)
sns.boxplot(x="day", y="total_bill", data=tips, notch=True)
````

````python
# 🔸 Use palette
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
````

# Violin Plot
➤ Used to show the distribution, probability density, central tendency, and spread of a numerical variable across categories.
```
🔧 1. Basic Syntax:sns.violinplot(x=None, y=None, data=None, hue=None, ...)
```
| Parameter | Description                                                              |
| --------- | ------------------------------------------------------------------------ |
| `x`       | Categorical variable (used for grouping)                                 |
| `y`       | Numerical variable (used for plotting)                                   |
| `data`    | DataFrame containing variables                                           |
| `hue`     | Subgrouping within each category                                         |
| `split`   | If `hue` is used, `split=True` puts both violins side-by-side inside one |
| `inner`   | What to draw inside (e.g., `'box'`, `'quartile'`, `'point'`, `None`)     |
| `bw`      | Bandwidth of KDE (smoothing). Lower = more jagged.                       |
| `scale`   | How the violin width is scaled: `'area'`, `'count'`, `'width'`           |
| `palette` | Color palette                                                            |
| `cut`     | Extends KDE beyond extreme data points (default: `2`)                    |

### 🔍 4. Interpretation
Width of the violin: Indicates density (how many points fall there)

Thicker areas: More observations

Thin ends: Fewer data points

White dot: Median (default when inner="box")

Box inside violin: Interquartile range

Lines inside: Quartiles (when inner='quartile')



````python
# Basic Example
sns.violinplot(x="smoker", y="tip", data=tips)
````

````python
# 🎨 5. CUSTOMIZATION
##🔸 Add Hue (with optional split)
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True)
````

````python
#  Draw only quartile lines inside
sns.violinplot(x="day", y="total_bill", data=tips, inner="quartile")
````

````python
# 🔸 Remove all inner markings
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
````

````python
# 🔸 Show individual points using swarmplot
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips, color="k", alpha=0.5)
````

````python
# Change of scale
sns.violinplot(x="day", y="total_bill", data=tips, scale="count")  # Width ∝ sample size
````

# Count Plot
✅ Purpose:

➤ Used to show the number of observations in each category of a categorical feature.
 showing counts of each category (like histogram for categorical data).

```
sns.countplot(x=None, y=None, data=None, hue=None, ...)

```
| Parameter    | Description                                                   |
| ------------ | ------------------------------------------------------------- |
| `x` / `y`    | Categorical variable (x = vertical bars, y = horizontal bars) |
| `data`       | DataFrame                                                     |
| `hue`        | Subgrouping by another categorical column                     |
| `order`      | Custom order of categories                                    |
| `palette`    | Colors                                                        |
| `orient`     | `'v'` (default) or `'h'`                                      |
| `saturation` | Transparency of bars (0 → transparent, 1 → solid)             |
| `linewidth`  | Thickness of bar borders                                      |


````python
# Basic Example
sns.countplot(x="day", data=tips)
plt.title("Number of Records per Day")
plt.show()
````

````python
# 🎨 4. Customizations
# 🔸 Add Hue (subgroup by gender, for example)
sns.countplot(x="day", hue="sex", data=tips)
````

````python
# 🔸 Horizontal Bars
sns.countplot(y="day", data=tips)
````

````python
# 🔸 Custom Order
sns.countplot(x="day", data=tips, order=["Thur", "sat", "sun", "friday"])
````

````python
# 🔸 Change Color Palette
sns.countplot(x="day", data=tips, palette="Set2")
````

````python
# 🔸 Add Counts as Labels
ax = sns.countplot(x="day", data=tips)
for p in ax.patches:
    ax.annotate(f"{p.get_height()}", (p.get_x() + 0.3, p.get_height() + 1))
````

# 🎯 Plot Decision Guide for Univariate + Categorical Analysis
| Scenario                                                                              | Best Plot                       | Why Use It?                                             | Notes                                            |
| ------------------------------------------------------------------------------------- | ------------------------------- | ------------------------------------------------------- | ------------------------------------------------ |
| ✅ You want to **count how many times each category occurs**                           | `countplot()`                   | Simple bar chart of category frequencies                | Most common first step                           |
| ✅ You want to **compare mean/median of a numerical variable across categories**       | `barplot()`                     | Aggregates numeric value (like avg total\_bill per day) | Uses `estimator` like `np.mean`, `np.median`     |
| ✅ You want to **visualize distribution + outliers** of numeric data per category      | `boxplot()`                     | Shows median, IQR, whiskers, and outliers               | Great for spotting spread & skewness             |
| ✅ You want to **visualize full distribution shape** (with density bumps) per category | `violinplot()`                  | KDE + boxplot hybrid                                    | Better than boxplot for non-normal distributions |
| ✅ You want to **see individual data points** along with category                      | `swarmplot()`                   | Shows each point with slight jitter                     | Good for small datasets                          |
| ✅ Same as swarmplot, but for **larger datasets**                                      | `stripplot()`                   | Similar to swarmplot but can overlap                    | Use `jitter=True`                                |
| ✅ You want to **compare subgroups within categories** (like male/female within day)   | Add `hue=` to any plot          | Adds layered comparison                                 | Works with countplot, boxplot, violin, barplot   |
| ✅ You want to **see proportions / percentages**                                       | Pie chart / normalized bar plot | Not in Seaborn directly                                 | Use Pandas or custom bar chart                   |


# 🧠 Use This Simple Rule of Thumb:
📊 Category Only → Use:
    
✅ countplot() — for frequency

✅ Pie/Donut chart — for proportions (use sparingly)

📈 Category vs Numeric → Use:
    
✅ barplot() — for aggregates

✅ boxplot() — for spread + outliers

✅ violinplot() — for distribution shape

✅ swarmplot() / stripplot() — for raw points

# 📌 Example Scenarios
| Question                                                    | Ideal Plot                            |
| ----------------------------------------------------------- | ------------------------------------- |
| How many times does each `day` appear?                      | `countplot(x="day")`                  |
| What’s the average `total_bill` per `day`?                  | `barplot(x="day", y="total_bill")`    |
| What’s the distribution of `total_bill` per `day`?          | `boxplot(x="day", y="total_bill")`    |
| Are there outliers in `tip` values for different `sex`?     | `boxplot(x="sex", y="tip")`           |
| How do `total_bill` values spread out per `day` with shape? | `violinplot(x="day", y="total_bill")` |
| See all actual `tip` values per `day`?                      | `swarmplot(x="day", y="tip")`         |

# Faceting
Faceting = Splitting data into multiple subplots based on one or more features.
It helps to compare distributions/relationships across categories.

| Tool        | Use Case                                      | Works With                  |
| ----------- | --------------------------------------------- | --------------------------- |
| `FacetGrid` | Most customizable subplot engine              | Any kind of plot (`.map()`) |
| `catplot`   | Faceted version of categorical plots          | bar, box, strip, etc.       |
| `relplot`   | Faceted version of relational plots           | scatter, line               |
| `PairGrid`  | Multi-feature pairwise grid with full control | Numeric datasets            |


## FacetGrid() — General-Purpose Faceting -->
```
 ✅ Syntax: g = sns.FacetGrid(data, col="col_var", row="row_var")
            g.map(sns.histplot, "x_column")
```

````python
# Example 
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

g = sns.FacetGrid(tips, col="sex", row="time")
g.map(sns.histplot, "total_bill", bins=10)
plt.show()
````

## catplot() — Faceted Categorical Plot (Auto Uses FacetGrid)
```
syntax: sns.catplot(data=..., x=..., y=..., kind="box", col="...", row="...")
```

````python
sns.catplot(data=tips, kind="box", x="day", y="total_bill", col="sex", row="time")
plt.show()
````

# Array Operations

```python
a1= np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(3,4)
print(a1)
print(a2)
```

```python
#scalar Operation

#Arithmetic Operation
print(a1*2)
#All the elements are ,multiplied by 2
print(a1**2)
```

```python
#Relational Operator
print(a1==15)
# Every element of the array will be checked by 15 if any element is equal to the 15 then for that True 
# Otherwise False will be returned 
print(a1>15)
print(a1<15)
# basically ,condition will bechecked for every numbers for those he condition hold true true will returened ,
#Otherwise false will be returned
```

```python
#vector Operations
#Arithmetic 
print(a1+a2)
#Since Shape of both array is same item adds element wise
```

# Array function


```python
a1 = np.random.random((4,4))
a1=np.round(a1*100)
print(a1)
```

```python
#max/min/Sum/Prod
print(np.max(a1))
print(np.min(a1))
print(np.sum(a1))
print(np.prod(a1))
# 0->Coloumn 1->Row
print(np.max(a1,axis=0))
# It gives maximum element in each coloumn 
print(np.max(a1,axis=1))
#It gives maximum element in each row

```

```python
#mean/median/std/var
print(np.mean(a1))
print(np.mean(a1,axis=1))
#mean Row wise
print(np.mean(a1,axis=0))
#Coloumn wise mean
#median kya deta hai , list ko sort karne ke bad beech wala (middle element) ko deta hai 
print(np.median(a1,axis=1))
print(np.std(a1,axis=1))
print(np.var(a1,axis=1))
```

```python
#Stacking 

#Stacking wwe can only dp when shape is same 
#Horizantal stacking
x1 = np.arange(12).reshape(3,4)
x2 = np.arange(12,24).reshape(3,4)
x3=np.hstack((x1,x2))# horizantal stacking
x4=np.vstack((x1,x2))
# we can stack as many arrays  as we want 
print(x3)

print(x3.shape)
print(x4)
print(x4.shape)
```

```python
#Splitting
print(np.hsplit(x3,2))
print("----------")
print(np.hsplit(x3,4))
print("-------------")
print(np.vsplit(x4,2))
```

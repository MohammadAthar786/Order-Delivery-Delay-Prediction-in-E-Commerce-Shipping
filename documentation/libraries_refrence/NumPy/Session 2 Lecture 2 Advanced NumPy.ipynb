```python
# Booolean Indexing
z= np.random.randint(1,100,24).reshape(6,4)
print(z)
```

```python
# find all numbers greater than 50
print(z>50)
print(z[z>50])
```

```python
# find all numbers which are  even 
print(z[z%2==0])
```

```python
# Find all numbers greater than 50 and even 
print(z[(z>50) & (z&2==0)]) # (&) bitwise and , we will not use logical  'and'  operator
```

```python
#Find all numvbers  divisible by 7
print(z[z%7!=0])
```

# Broadcasting 

The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.
The smaller array is "broadcast- across the larger array so that they have compatible shapes.



1. Make the two arrays have the same number of dimensions..
• If the numbers of dimensions of the two arrays are different. add new dimensions with size 1 to the head of the array with the
smaller dimension.
2. Make each dimension of the two arrays the same size.
• If the sizes of each dimension of the two arrays do not match. dimensions with size 1 are stretched to the size of the other
array.
• If there is a dimension whose size is not 1 in either of the two arrays. it cannot be broadcasted, and an error is raised.

```python
#Mean Square error 

def MeanSE(actual,predicted):
        return np.mean((actual-predicted)**2)
    
    

actual =np.random.randint(1,50,25)
print(actual)
predicted=np.random.randint(1,50,25)
print(predicted)
print(MeanSE(actual,predicted))
```

# Binary Cross  Entropy



```python
def BinaryCrossEntropy(actual,predicted):
    return np.mean(actual*(np.log(predicted))+(1-actual)*(np.log(1-predicted)))

actual=np.arange(12)
predicted=np.arange(12,24)
print(BinaryCrossEntropy(actual,predicted))
```

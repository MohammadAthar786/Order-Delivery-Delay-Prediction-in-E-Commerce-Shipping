```python
import matplotlib.pyplot as plt
```

```python
# Examples 
x = [1, 2, 3]
y = [4, 5, 6]

plt.plot(x, y)  # Line plot
plt.show()

plt.scatter(x, y)  # Scatter plot
plt.show()

plt.bar(['A', 'B', 'C'], [3, 5, 7])  # Bar chart
plt.show()

import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=10)  # Histogram
plt.show()

sizes = [30, 40, 30]
labels = ['A', 'B', 'C']
plt.pie(sizes, labels=labels)  # Pie chart
plt.show()
```

```python
# Examples 
plt.plot(x, y)
plt.title('Sales Data')  # Title
plt.xlabel('Days')  # X-axis label
plt.ylabel('Sales')  # Y-axis label
plt.legend(['Product A'])  # Legend
plt.grid(True)  # Show grid
plt.xticks([1, 2, 3], ['Jan', 'Feb', 'Mar'])  # Custom X-axis labels
plt.yticks([4, 5, 6], ['Low', 'Medium', 'High'])  # Custom Y-axis labels
plt.show()
```

```python
# Examples
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2, marker='o')  
plt.show()
```

```python
# Multiple Plots
x1 = [1, 2, 3]
y1 = [4, 5, 6]

x2 = [1, 2, 3]
y2 = [6, 5, 4]

plt.plot(x1, y1, label='Line 1')
plt.plot(x2, y2, label='Line 2')
plt.legend()
plt.show()
```

```python
# Examples 
plt.subplot(2, 2, 1)  # 2x2 grid, first subplot
plt.plot(x, y)

plt.subplot(2, 2, 2)  # Second subplot
plt.bar(['A', 'B', 'C'], [3, 5, 7])

plt.subplot(2, 2, 3)  # Third subplot
plt.scatter(x, y)

plt.subplot(2, 2, 4)  # Fourth subplot
plt.hist(data, bins=10)

plt.show()
```

```python
# Examples 
plt.bar(['A', 'B', 'C'], [10, 20, 15])  # Vertical bars
plt.show()

plt.barh(['A', 'B', 'C'], [10, 20, 15])  # Horizontal bars
plt.show()
```

```python
# Examples 
plt.scatter(x, y, color='red', s=100)  # Scatter with size and color
plt.show()
```

```python
# Examples 
plt.hist(data, bins=10, color='green')  # Green histogram
plt.show()
```

```python
# Examples 
plt.pie([30, 40, 30], labels=['A', 'B', 'C'], autopct='%1.1f%%', explode=[0, 0.1, 0])
plt.show()
```

```python
# Examples 
plt.plot(x, y)
plt.savefig('chart.png')  # Save as PNG
plt.savefig('chart.pdf')  # Save as PDF
plt.show()
```

```python
# Examples 
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Sales Data')  
ax.set_xlabel('Months')  
ax.set_ylabel('Revenue')  
plt.show()
```

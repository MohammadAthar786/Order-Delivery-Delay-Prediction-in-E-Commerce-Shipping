```python
import os
import pandas as pd
from sqlalchemy import  types

# 🧠 Schema definitions (IMPORTANT 🔥)
schemas = {

    "orders": {
        "order_id": types.VARCHAR(50),
        "customer_id": types.VARCHAR(50),
        "order_status": types.VARCHAR(20),
        "order_purchase_timestamp": types.TIMESTAMP(),
        "order_approved_at": types.TIMESTAMP(),
        "order_delivered_carrier_date": types.TIMESTAMP(),
        "order_delivered_customer_date": types.TIMESTAMP(),
        "order_estimated_delivery_date": types.DATE()
    },

    "order_items": {
        "order_id": types.VARCHAR(50),
        "order_item_id": types.INT(),
        "product_id": types.VARCHAR(50),
        "seller_id": types.VARCHAR(50),
        "shipping_limit_date": types.TIMESTAMP(), # In MySQL function is DATETIME()
        "price": types.DECIMAL(10,2),
        "freight_value": types.DECIMAL(10,2)
    },

    "customers": {
        "customer_id": types.VARCHAR(50),
        "customer_unique_id": types.VARCHAR(50),
        "customer_zip_code_prefix": types.INT(),
        "customer_city": types.VARCHAR(50),
        "customer_state": types.CHAR(2)
    },

    "order_payments": {
        "order_id": types.VARCHAR(50),
        "payment_sequential": types.INT(),
        "payment_type": types.VARCHAR(20),
        "payment_installments": types.INT(),
        "payment_value": types.DECIMAL(10,2)
    },

    "order_reviews": {
        "review_id": types.VARCHAR(50),
        "order_id": types.VARCHAR(50),
        "review_score": types.INT(),
        "review_comment_title": types.TEXT(),
        "review_comment_message": types.TEXT(),
        "review_creation_date": types.TIMESTAMP(),   # In MySQL the function is DATETIME 
        "review_answer_timestamp": types.TIMESTAMP()
    },

    "products": {
        "product_id": types.VARCHAR(50),
        "product_category_name": types.VARCHAR(100),
        "product_name_length": types.INT(),
        "product_description_length": types.INT(),
        "product_photos_qty": types.INT(),
        "product_weight_g": types.INT(),
        "product_length_cm": types.INT(),
        "product_height_cm": types.INT(),
        "product_width_cm": types.INT()
    },

    "sellers": {
        "seller_id": types.VARCHAR(50),
        "seller_zip_code_prefix": types.INT(),
        "seller_city": types.VARCHAR(50),
        "seller_state": types.CHAR(2)
    }
}



```

```python
## Loading script


from pathlib import Path
import pandas as pd
from db_connection import engine
import os
DATA_DIR = r"C:\Users\mohda\OneDrive\Desktop\Olist Project\data"

for file in os.listdir(DATA_DIR):
    if file.endswith(".csv"):
        
        table_name = file.replace(".csv", "")
        file_path = os.path.join(DATA_DIR, file)

        print(f"🚀 Loading {table_name}")

        df = pd.read_csv(file_path, low_memory=False)

        df.to_sql(table_name, con=engine, if_exists='replace', index=False, dtype=schemas.get(table_name, None))

print("✅ ALL TABLES LOADED WITH CLEAN SCHEMA")

print("All CSV files loaded successfully.")
```
